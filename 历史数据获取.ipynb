```python
df = pro.query('daily_basic', ts_code='', trade_date='20251111',fields='ts_code,trade_date,turnover_rate,total_share,float_share,free_share,total_mv,circ_mv')
```

```python
stock_list = df['ts_code'].to_list()
```

```python
i = 1
for codename in stock_list:
    try:
        tmpdf1 = pro.query('daily', ts_code=codename, end_date='20251111')
        tmpdf1 = tmpdf1[['ts_code','trade_date','open','close','low','high','vol','amount']]
        tmpdf1.columns = ['code','trade_datetime','open','close','low','high','volume','amount']
        tmpdf1 = tmpdf1.copy()
        tmpdf1['trade_datetime'] = (
            pd.to_datetime(tmpdf1['trade_datetime'], format='%Y%m%d')   # 20251111 → 2025-11-11 00:00:00
            + pd.Timedelta(hours=15)                            # 加上 15 小时
        )
        tmpdf1 = tmpdf1.sort_values('trade_datetime').reset_index(drop=True)
    
        filename = codename + '.csv'
        tmpdf1.to_csv(filename, index=False)
    except Exception as e:
        print(f'{codename}获取失败，原因{e}')
    
    i = i + 1
    if i == 500:
        print('休息1分钟')
        time.sleep(20)
        i = 1
```
